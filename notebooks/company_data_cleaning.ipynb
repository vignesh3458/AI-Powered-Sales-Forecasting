```python
import pandas as pd

df = pd.read_excel(
    "../company_dataset/Restaurant_Sales_Sample_Data.xlsx"
)

print(df.shape)
print(df.info())

df.to_csv(
    "../cleaned_data/restaurant_sales.csv",
    index=False
)

print("Dataset Saved Successfully")
```

```python
print("Duplicates:", df.duplicated().sum())

df = df.drop_duplicates()

print("After Cleaning:", df.shape)
```

```python
df["Date"] = pd.to_datetime(df["Date"])

df.info()
```

```python
for date in df["Date"]:
    try:
        pd.to_datetime(date)
    except Exception:
        print(date)
```

```python
df["Date"] = pd.to_datetime(
    df["Date"],
    dayfirst=True,
    errors="coerce"
)
```

```python
df["Date"].isnull().sum()
```

```python
df["Day"] = df["Date"].dt.day

df["Week"] = df["Date"].dt.isocalendar().week

df["YearMonth"] = df["Date"].dt.strftime("%Y-%m")
```

```python
df.head()
```

```python
df.to_csv(
    "../cleaned_data/restaurant_sales_cleaned.csv",
    index=False
)

print("Dataset Saved Successfully")
```
